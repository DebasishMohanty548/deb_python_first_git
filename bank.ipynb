```python
import pandas as pd
import matplotlib.pyplot as mat
import seaborn as sea
import numpy as np
```

```python
path=r"C:\Users\debasish\DS-AI\N-IT\datafiles\bank.csv"
bank=pd.read_csv(path)
bank.dtypes
```

```python
d=dict(bank.dtypes)
obj=[i for i,v in d.items() if v=='object']
for i in obj:
    cd=bank[i].value_counts()
    keys=cd.keys()
    val=cd.values
    bankframe=pd.DataFrame(zip(keys,val),columns=['label','counts'])
    mat.figure(figsize=(17,3))
    mat.bar('label','counts',data=bankframe)
    mat.xlabel('label')
    mat.ylabel('count')
    mat.title(f"{i} data")
    mat.show()
```

```python
d=pd.DataFrame()
for i in num:
    mean=round(bank[i].mean(),2)
    med=round(bank[i].median(),2)
    std=round(bank[i].std(),2)
    c=len(bank[i])
    mx=round(max(bank[i]),2)
    mn=round(min(bank[i]),2)
    p75=round(np.percentile(bank[i],75),2)
    p50=round(np.percentile(bank[i],50),2)
    p25=round(np.percentile(bank[i],25),2)
    pli=[mean,med,std,c,mx,mn,p25,p50,p75]
    index=['mean','median','std','count','max','min','25p','50p','75p']
    wage_dt=pd.DataFrame(pli,index=index,columns=[i])
    d=pd.concat([d,wage_dt],axis=1)
```
