```python
class Pattern:
    def __init__(self, n,t):
        self.n = n
        self.t=t
    def print_pattern(self, char_func, space_func, end_val_func):
        k = char_func()
        for i in range(1, self.n + 1):
            for j in range((self.n - i) + 1):
                print(end=space_func())
            for j in range(1, end_val_func(i) + 1):
                print(chr(k), end=' ')
                k += 1
            print()

    def left_right_tri(self, upper=True, even=False, odd=False):
        char_func = lambda: ord('A') if upper else ord('a')
        space_func = lambda: '' if even or odd else ' '
        end_val_func = lambda i: 2 * i if even or odd else i
        self.print_pattern(char_func, space_func, end_val_func)

    def right_right_tri(self, upper=True, even=False, odd=False):
        char_func = lambda: ord('A') if upper else ord('a')
        space_func = lambda: '  '
        end_val_func = lambda i: 2 * i if even or odd else i
        self.print_pattern(char_func, space_func, end_val_func)

    def equilateral_tri(self, upper=True, even=False, odd=False):
        char_func = lambda: ord('A') if upper else ord('a')
        space_func = lambda: ' '
        end_val_func = lambda i: 2 * i if even or odd else i
        self.print_pattern(char_func, space_func, end_val_func)

    def print_symbol_pattern(self, pattern_func, even=False, odd=False):
        for i in range(1, self.n + 1):
            for j in range((self.n - i) + 1):
                print(end='  ' if even or odd else ' ')
            for j in range(1, (2 * i + 1) if even or odd else i + 1):
                print(self.t, end=' ')
            print()

    def symbol_pattern(self, pattern_type, even=False, odd=False):
        if pattern_type == 1:
            self.print_symbol_pattern(self.equilateral_tri, even, odd)
        elif pattern_type == 2:
            self.print_symbol_pattern(self.right_right_tri, even, odd)
        elif pattern_type == 3:
            self.print_symbol_pattern(self.left_right_tri, even, odd)


n = int(input('Enter num: '))
t= input("enter symbol : ")
p = Pattern(n,t)
print("---PATTERNS DISPLAY---")
c = input('Enter c if you want character pattern else enter s for symbol pattern: ')

if c.lower() == 'c':
    case = input('If you want alphabet in uppercase enter u else enter l: ').lower()
    normal = input("Do you want to display in odd/even sequence? Enter y/n: ").lower()
    oe = input('Enter o for odd numbers or e for even numbers: ').lower() if normal == 'y' else None

    upper = case == 'u'
    even = oe == 'e' if oe else False
    odd = oe == 'o' if oe else False

    j = int(input("Enter 1 for EQUILATERAL TRIANGLE \nEnter 2 for r_RIGHT ANGLE TRIANGLE \nEnter 3 for l_RIGHT ANGLE TRIANGLE \nEnter 4 for exit: "))

    if j == 1:
        p.equilateral_tri(upper, even, odd)
    elif j == 2:
        p.right_right_tri(upper, even, odd)
    elif j == 3:
        p.left_right_tri(upper, even, odd)
    elif j == 4:
        exit()
    else:
        print("Invalid choice")
        exit()
elif c.lower() == 's':
    normal = input("Do you want to display in odd/even sequence? Enter y/n: ").lower()
    oe = input('Enter o for odd numbers or e for even numbers: ').lower() if normal == 'y' else None

    even = oe == 'e' if oe else False
    odd = oe == 'o' if oe else False

    j = int(input("Enter 1 for EQUILATERAL TRIANGLE \nEnter 2 for r_RIGHT ANGLE TRIANGLE \nEnter 3 for l_RIGHT ANGLE TRIANGLE \nEnter 4 for exit: "))

    if j in [1, 2, 3]:
        p.symbol_pattern(j, even, odd)
    elif j == 4:
        exit()
    else:
        print("Invalid choice")
        exit()
else:
    print('Invalid input')
```
